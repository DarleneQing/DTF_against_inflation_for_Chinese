```python
import pandas as pd
```

```python
df = pd.read_excel("../data/data.xlsx")
df
```

```python
# formatt date time and seperate year and quarter

df['Time']=pd.to_datetime(df['Time'])
df['Year']=df['Time'].dt.year
df['Quarter']=df['Time'].dt.quarter
df['Quarter'] = 'Q' + df['Quarter'].astype(str)
```

```python
# Convert label assignments to column descriptions

df['House_price'].attrs['description'] = "(CNY/m^2)"
df['Gold_price'].attrs['description'] = "(CNY/gram)"
df['Fixed_deposit_rate'].attrs['description'] = "percentage%"
df['Shanghai_stock_index'].attrs['description'] = "point"
df['government_bond_yield_sixm'].attrs['description'] = "government_bond_yield(6-month)"
df['government_bond_yield_fivey'].attrs['description'] = "government_bond_yield(5-year)"
df['government_bond_yield_teny'].attrs['description'] = "government_bond_yield(10-year)"
```

```python
# first percentage change on house, gold, deposite and stock
first_pct = ['House_price', 'Gold_price', 'Shanghai_stock_index']
df[first_pct] = df[first_pct].pct_change() * 100

# Replace NA with second row as default value
for col in first_pct:
    if pd.isna(df.at[0, col]):
        df.at[0, col] = df.at[1, col]

df
```

```python
# create pickle file for original data exclude datetime
data = df.drop(columns=['Year','Quarter']).copy()
data.to_pickle('../data/data_only.pkl')
```

```python
# create pickle files of percentage change for assets and inflation indicators

df_pct = df.drop(columns=['Time','Year','Quarter']).copy()
df_pct = df_pct.pct_change() * 100
df_pct.columns = ['House_Change', 'Gold_Change', 'Fixed_deposit_Change', 
      'Shanghai_stock_Change', 'Core_CPI_Change','Headline_CPI_Change','Core_inflation_Change','Headline_inflation_Change', 'Gover_sixm_Change', 'Gover_fivey_Change', 
      'Gover_teny_Change']

# Replace NA with second row as default value
second_row_values = df_pct.iloc[1]  # Get values from the second row
df_pct.fillna(second_row_values, inplace=True)
    

pct_asset = df_pct.drop(columns=['Core_CPI_Change','Headline_CPI_Change','Core_inflation_Change','Headline_inflation_Change']).copy()
pct_asset[['Time','Year','Quarter']] = df[['Time','Year','Quarter']].copy()
pct_asset.to_pickle('../data/pct_asset.pkl')

pct_inflation = df_pct[['Core_CPI_Change','Headline_CPI_Change','Core_inflation_Change','Headline_inflation_Change']].copy()
pct_inflation[['Time','Year','Quarter']] = df[['Time','Year','Quarter']].copy()
pct_inflation.to_pickle('../data/pct_inflation.pkl')
```

```python
# create a subset for all the assets only and save it as pickle

df_asset_subset = df.drop(columns=['core_CPI','headline_CPI','core_inflation','headline_inflation']).copy()
df_asset_subset.to_pickle('../data/asset_subset.pkl')

# Also for inflation indicators
df_inflation_subset = df[['Time','Year', 'Quarter','core_CPI','headline_CPI','core_inflation','headline_inflation']].copy()
df_inflation_subset.to_pickle('../data/inflation_subset.pkl')
```
